```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Headers
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

baseurl = "https://loveandflair.com"

# -------------------------------
# Step 1: Get all product links
# -------------------------------
productlinks = []

for x in range(1, 12):
    url = f"{baseurl}/collections/activewear-1?page={x}"
    print(f"Fetching page {x}...")

    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.content, "lxml")

    for a in soup.select('a[href*="/products/"]'):
        link = a.get("href")

        if link:
            full_link = baseurl + link

            if full_link not in productlinks:
                productlinks.append(full_link)

    time.sleep(1)  # polite delay

print("Total product links:", len(productlinks))


# -------------------------------
# Step 2: Scrape each product
# -------------------------------
all_data = []

for link in productlinks:
    print("Scraping:", link)

    try:
        r = requests.get(link, headers=headers)
        soup = BeautifulSoup(r.content, "lxml")

        # Name
        name_tag = soup.find('h4', class_="product__title")
        name = name_tag.text.strip() if name_tag else None

        # Price
        price_tag = soup.find('span', class_="money")
        price = price_tag.text.strip() if price_tag else None

        # Stock
        stock_tag = soup.find('span', class_="advantage__title")
        stock = stock_tag.text.strip() if stock_tag else None

        # Colors
        colors = [
            c.get("value")
            for c in soup.select('input[type="radio"][name*="Color"]')
        ]

        # Sizes
        sizes = [
            s.get("value")
            for s in soup.select('input[type="radio"][name*="Size"]')
        ]

        # Material
        desc = soup.find("div", class_="about__accordion-description")

        material = None

        if desc:
            all_texts = [t.strip() for t in desc.stripped_strings]

            for i in range(len(all_texts)):
                if all_texts[i] == "MATERIAL":
                    # next non-empty text = material
                    for j in range(i + 1, len(all_texts)):
                        if all_texts[j]:
                            material = all_texts[j]
                            break
                    break

  

        # Store data
        all_data.append({
            "Name": name,
            "Price": price,
            "Stock": stock,
            "Colors": ", ".join(colors),
            "Sizes": ", ".join(sizes),
            "Material": material,
            "Link": link
        })

        time.sleep(1)  # avoid blocking

    except Exception as e:
        print("Error:", e)
        continue


print("Total products scraped:", len(all_data))


# -------------------------------
# Step 3: Save to CSV
# -------------------------------
df = pd.DataFrame(all_data)
df.to_csv("loveandflair_products.csv", index=False)

print("Data saved to loveandflair_products.csv")
```
